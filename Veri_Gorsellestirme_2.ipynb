```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Öğrenci sayısı => (Number of students)
ogrenci_sayisi = 30

# Şube 1'in ders notları => (Lecture notes of Branch 1)
s1_ders1_notlar = [75, 80, 85, 90]
s1_ders2_notlar = [70, 82, 88, 95]
s1_ders3_notlar = [65, 78, 80, 92]
s1_ders4_notlar = [72, 85, 90, 88]

# Şube 2'nin ders notları => (Lecture notes of Branch 2)
s2_ders1_notlar = [78, 83, 88, 91]
s2_ders2_notlar = [68, 79, 84, 89]
s2_ders3_notlar = [71, 80, 86, 90]
s2_ders4_notlar = [75, 82, 87, 92]

# Ortalama hesaplama => (Average calculation)
s1_ders1_ortalama = sum(s1_ders1_notlar) / ogrenci_sayisi
s1_ders2_ortalama = sum(s1_ders2_notlar) / ogrenci_sayisi
s1_ders3_ortalama = sum(s1_ders3_notlar) / ogrenci_sayisi
s1_ders4_ortalama = sum(s1_ders4_notlar) / ogrenci_sayisi

s2_ders1_ortalama = sum(s2_ders1_notlar) / ogrenci_sayisi
s2_ders2_ortalama = sum(s2_ders2_notlar) / ogrenci_sayisi
s2_ders3_ortalama = sum(s2_ders3_notlar) / ogrenci_sayisi
s2_ders4_ortalama = sum(s2_ders4_notlar) / ogrenci_sayisi

# Bar grafikleri için verileri oluşturma => (Create data for bar charts)
ders_ortalama_s1 = [s1_ders1_ortalama, s1_ders2_ortalama, s1_ders3_ortalama, s1_ders4_ortalama]
ders_ortalama_s2 = [s2_ders1_ortalama, s2_ders2_ortalama, s2_ders3_ortalama, s2_ders4_ortalama]
dersler = ['Ders 1', 'Ders 2', 'Ders 3', 'Ders 4']

x = np.arange(len(dersler))  # x ekseninde derslerin konumunu belirleme => (# Determining the order of x lessons)
bar_genisligi = 0.35

# Bar grafiklerini çizme => (Drawing bar charts)
plt.bar(x, ders_ortalama_s1, width=bar_genisligi, label='Şube 1')
plt.bar(x + bar_genisligi, ders_ortalama_s2, width=bar_genisligi, label='Şube 2')
plt.xlabel('Dersler')
plt.ylabel('Ortalama Notlar')
plt.title('Her Bir Ders İçin Şube 1 ve Şube 2 Ortalama Notları')
plt.xticks(x + bar_genisligi/2, dersler)
plt.legend()

# Grafiği gösterme => (Show chart)
plt.show()
```

```python
np.random.seed(0)  # Rastgele sayı üretimi için sabit bir başlangıç noktası belirleyelim => (Let's set a fixed starting point for random number generation)

# 4 sınıfın 4 sınavdaki notlarını oluşturalım => (Let's create the scores of 4 classes in 4 exams)
notlar = np.random.randint(0, 101, size=(4, 4, 30))  # 4 sınıf, 4 sınav, 30 öğrenci

# Sınav notlarının ortalamalarını alalım => (Let's take the average of the exam scores)
sınav_ortalamaları = np.mean(notlar, axis=2)  # Her bir sınavın ortalaması için son eksen boyunca ortalama alınır => (The average is taken along the last axis as each exam is taken)

# Her bir sınavın bar grafiğini çizdirme => (Draw bar charts for each exam)
fig, ax = plt.subplots(1, 4, figsize=(20, 5))  # 1 satır, 4 sütunlu bir subplot oluşturalım => (Let's create a 1 row, 4 column subplot)
sınavlar = ["Sınav 1", "Sınav 2", "Sınav 3", "Sınav 4"]

for i in range(4):
    ax[i].bar(np.arange(4), sınav_ortalamaları[i], color='skyblue')  # Her bir sınavın ortalamasını çizdirme => (Plot the average of each exam)
    ax[i].set_xticks(np.arange(4))  # x ekseni için etiketleri belirleme => (Specify labels for the x-axis)
    ax[i].set_xticklabels(["Şube 1", "Şube 2", "Şube 3", "Şube 4"])  # x ekseni etiketlerini belirleme => (Specify labels for the x-axis)
    ax[i].set_title(sınavlar[i])  # Grafik başlığını belirleme => (Determine chart titles)

plt.tight_layout()  # Grafiği sıkıştırma => (Compress chart)
plt.show()  # Grafiği gösterme => (Show chart)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Verileri oluşturma => (Create the data)
notlar = np.random.randint(0, 100, (30, 4))  # 4 sınıfın 30 öğrencisinin notları => (Grades of 30 students of 4 classes)

# Her sınıfın ortalamasını hesapla => (Calculate the average of each class)
ortalama_notlar = np.mean(notlar, axis=0)

# Sınıf etiketleri => (Class labels)
siniflar = ['A', 'B', 'C', 'D']

# Renkler => (colors)
renkler = ['blue', 'green', 'red', 'orange']

# Bar plotları yan yana çizme => (Drawing bar plots side by side)
bars = plt.bar(siniflar, ortalama_notlar, color=renkler)

plt.xlabel('Sınıflar')
plt.ylabel('Ortalama Notlar')
plt.title('Sınıf Ortalama Notları')

# Legend (Leyenda)
plt.legend(bars, siniflar)

for i in range(len(siniflar)):
    plt.text(i,ortalama_notlar[i],f"{ortalama_notlar[i]:.2f}",ha="center",va="bottom")
plt.show()
```
