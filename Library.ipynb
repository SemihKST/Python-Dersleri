```python
my_arr1 = np.random.randint(20,50,(5))
my_arr2 = np.random.randint(50,100,(5))

my_dizi_isim = ["Ahmet","Mehmet","Furkan","Yusuf","Hakan"]

my_dict_1 = {"isim":my_dizi_isim,"yas":my_arr1}
my_dict_2 = {"isim":my_dizi_isim,"kilo":my_arr2}

my_df_1 = pd.DataFrame.from_dict(my_dict_1)
my_df_2 = pd.DataFrame.from_dict(my_dict_2)
```
